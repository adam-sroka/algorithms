```python
import numpy as np
import matplotlib.pyplot as plt

# use retina plots
%config InlineBackend.figure_format='retina'
```

```python
# define dice
# assume that each dice has values 0..<d
d = 6
dice_prob = np.ones(d) / d

# initialize array of outcome probabilities
roll_prob = [dice_prob]

# convolve
n_rolls = 10
for i in range(1, n_rolls):
    # note: mode='full' corresponds to zero-padding
    roll_prob += [np.convolve(roll_prob[-1], dice_prob, mode='full')]
    
# plot outcome probabilities
fig, ax = plt.subplots()
for i, x in enumerate(roll_prob):
    ax.plot(x, '.-', label=f'{i+1:2} rolls')
ax.legend()
ax.set_xlabel('sum')
ax.set_ylabel('probability')
plt.show()
```
